```python
def plot_solution(x0,days,N,lkdwn_start,lkdwn_end,day,kt1,beta,beta1):
    
    data=pd.read_csv("maha_data7apr.csv")
    real_P=data["Total Cases"]-data["Total Deaths"]-data["Total Recoveries"]
    real_P=real_P.values
    real_t=list(range(0,len(real_P)))
    date=list(data["Date"])
    real_PP=list(real_P)
    real_PP.extend([962,1158,1293,1443])
    real_t.extend([25,26,27,28])
    
    def sirx(x,t):
            #import the values
            day1=list(map(int,day.split(',')))
            day1.insert(0,0)
            day1.append(days)
            kt2=list(map(float,kt1.split(',')))
            S,E,A,I,Xs,Xe,Xa,Xi,P,R=x

            #Parameters
            beta2=0.1
            b1=beta1
            b2=0.1*beta1
            b3=0.002*beta1
            k0=1/7
            kt=kt2[0]
            mu=1/7
            sigma=1/5
            gamma1=1/19
            gamma2=1/14
            gamma3=1/14
            f=0.1

            lkdwn_start1=[date.index(lkdwn_start)]
            lkdwn_end1=[lkdwn_end]
            lkdwn_start1.append(days)
            
            #Change Kt dates
            for i in range(len(day1)-1):
                if t>=day1[i] and t<day1[i+1]:
                    kt=kt2[i]
                    break
            
            #For heavyside function
            for j in range(len(lkdwn_end1)):
                if t>lkdwn_end1[j] and t<lkdwn_start1[j+1]:
                    theta=1
                    break
                else:
                    theta=0
            #For Pulse function
            for j in range(len(lkdwn_end1)):
                if t>lkdwn_start1[j] and t<lkdwn_end1[j]:
                    zi=1
                    break
                else:
                    zi=0

            com=beta*(I+A+(b1*Xa)+(b2*Xi)+(b3*P)+beta2*(E+beta1*Xe))
            if t%1==0:
                print(kt,t)
            #THE EQUATIONS
            dSdt = -(com*S/N) - (zi*k0*S) + (theta*mu*Xs)
            dEdt = (f*com*S/N) - (gamma1*E) - (kt*0.5*E)
            dAdt = ((1-f)*com*S/N) - (sigma*A) - (zi*k0*A) + (theta*mu*Xa) - (kt*0.5*A)
            dIdt = (sigma*A) - (kt*0.8*I) - (zi*k0*I) + (theta*mu*Xi) - (gamma2*I)                              
            dXsdt = -(beta1*com*Xs/N) + (zi*k0*S) - (theta*mu*Xs)
            dXedt = (f*beta1*com*Xs/N) - (gamma1*Xe) - (kt*0.5*Xe)
            dXadt = ((1-f)*beta1*com*Xs/N) - (sigma*Xa) + (zi*k0*A) - (theta*mu*Xa) - (kt*0.5*Xa)
            dXidt = (sigma*Xa) - (kt*0.8*Xi) + (zi*k0*I) - (theta*mu*Xi) - (gamma2*Xi)
            dPdt = (kt*(0.8*(I+Xi) + 0.5*(A+Xa+E+Xe))) - (gamma3*P)
            dRdt = (gamma2*(I+Xi)) + (gamma3*P) + gamma1*(E + Xe)

            return [dSdt,dEdt,dAdt,dIdt,dXsdt,dXedt,dXadt,dXidt,dPdt,dRdt]
    #Ploting the equations
    t=np.linspace(0,days,(days*1000))
    x=odeint(sirx,x0,t)
    S = x[:,0]
    E = x[:,1]
    A = x[:,2]
    I = x[:,3]
    Xs = x[:,4]
    Xe = x[:,5]
    Xa = x[:,6]
    Xi = x[:,7]
    P = x[:,8]
    R = x[:,9]
    
    arr=['Suceptible','Exposed','Asymptomatic','Infected','Quarantined_Suceptible','Quarantined_Exposed','Qurantined_Asymptomatic','Qurantined_Infected','Patients','Removed']
    #text="N={}, lkdwn_start={}, lkdwn_end={},\nbeta={}, beta1={}, b1={}, b2={}, b3={},\n k0={}, kt={}, mu={}, sigma={},\n gamma1={}, gamma2={}, gamma3={}".format(N ,lkdwn_start ,lkdwn_end ,beta ,beta1 ,b1 ,b2 ,b3 ,k0 ,kt ,mu ,sigma ,gamma1 ,gamma2 ,gamma3)
    '''
    plt.figure(figsize=(10,5))
    for i in range(len(x0)):
        plt.subplot(2,4,i+1)
        plt.subplots_adjust(hspace =1,wspace=0.5)
        plt.plot(t,(x[:,i]/N)*100)
        plt.ylabel(arr[i])
        plt.xlabel("Time in Days")
    plt.show()
    
    plt.figure(figsize=(20,10))
    plt.plot(t,(x/N)*100)
    plt.legend(arr)
    plt.xlabel("Time in Day"+"\n"+text)
    plt.show()
    '''
    plt.figure(figsize=(20,10))
    plt.plot(t,P,label='Simulation')
    plt.axvline(x=24,color='black')
    plt.scatter(real_t,real_PP,color='r',label='real P')
    plt.legend()
    #plt.xlabel("Time in Day"+"\n"+text)
    plt.show()

    '''plt.plot(t,R,label='Simulation')
    plt.plot(time,actualR,label='Actual')
    plt.legend()
    plt.xlabel("Time in Day"+"\n"+text)
    plt.show()'''
```
